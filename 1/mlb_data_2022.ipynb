```python
# since some players have played for multiple teams, we need to find all players with Tm == 'TOT' and remove all of their rows except the one with Tm == 'TOT'
# 
# we can do this by first finding all players with Tm == 'TOT' and then finding all rows with the same player name and removing them
# 


players_with_tot = df.query("Tm == 'TOT'")['Name'].unique()

# find all rows with the same player name and remove them (except the one with Tm == 'TOT')
print(len(players_with_tot))
df = df.query("not Name in @players_with_tot | Tm == 'TOT'")
```
